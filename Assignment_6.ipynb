```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("weather_3.csv")
```

```python
df=pd.DataFrame(data)
df.head(1)
```

```python
x=df["MaxTemp"]
sns.distplot(x)
```

```python
x=df["MinTemp"]
sns.distplot(x, kde=False, rug=True)
```

```python
x=df["Rainfall"]
sns.distplot(x, hist=False, rug=True)
```

```python
x=df["Evaporation"]
sns.kdeplot(x, shade=True)
```

```python
sns.jointplot(x='Evaporation', y='Sunshine',data = df)
```

```python
with sns.axes_style('white'):
    sns.jointplot(x = df['Pressure9am'], y = df['Temp9am'], kind = 'hex', color = 'k')
```

```python
sns.jointplot(x='Cloud9am', y='Humidity9am', data=df, kind='kde')
```

```python
plt.scatter(x=df["WindGustSpeed"], y=df["Sunshine"])
```

```python
df_1=pd.DataFrame(df, columns=['Humidity3pm','Humidity9am','Cloud3pm','Cloud9am','RainToday'])
sns.pairplot(df_1,hue='RainToday')
```

```python
sns.stripplot(x ='Cloud3pm' , y='Humidity3pm', data=df)
```

```python
sns.stripplot(x ='Cloud3pm' , y='Humidity3pm', data=df, jitter=True)
```

```python
sns.swarmplot(x ='Cloud3pm' , y='Humidity3pm', data=df)
```

```python
sns.boxplot(x ='Cloud9am' , y='Humidity9am', data=df)
```

```python
sns.boxplot(x ='Cloud9am' , y='Humidity9am', data=df,hue='RainToday', dodge=True)
```

```python
sns.barplot(x='RainToday', y='Cloud3pm', data=df)
```

```python
sns.barplot(x='Cloud3pm', y='Humidity3pm', data=df, hue='RainToday')
```

```python
sns.countplot(x='WindGustDir', data=df, palette='Greens_d')
```

```python
plt.figure(1)

plt.subplots_adjust(hspace=0.5 )

plt.subplot(211)
sns.countplot(x='RainTomorrow', data=df)

plt.subplot(212)
sns.countplot(x='WindGustDir', data=df)

plt.show()
```
